```
# add a column 'ProteinID', only one protein ID kept in the 'Protein IDs' column. Keep the one have the largest 'Peptide counts (all)'. If more than one proteins have the largest 'Peptide counts (all)' value, keep OGS2 ID
def getProteinID(r):
    '''
    r is one row of dataframe
    proteinIDs looks like 'Msex2.00001-RA;MCOT.C00001.1.0.TOO5O;MCOT.C00001.3.0.OOO5O;MCOT.C00001.2.0.TOO5O'
    keep Msex2 id if there is Msex2 id, otherwise, keep the first ID
    '''
    l = r['Peptide counts (all)'].split(';')
    l = [e for e in l if e == l[0]]
    n = r['Protein IDs'].split(';')
    prs = n[:len(l)]
    prs_keep = [e for e in prs if e.startswith('Msex2.')]
    if len(prs_keep) > 0:
        return prs_keep[0]
    return prs[0]
```

### calculate iBAQ/intensity for each sample

The conclusion is, yes, iBAQ = intensity/peptides, where peptides is the number of theoretical detectable peptides.

````python
# extract iBAQ and intensity
df_iBAQ = df.loc[:,[e for e in df.columns if 'iBAQ' in e]]#iBAQ columns
df_intensity = df.loc[:,[e for e in df.columns if 'Intensity' in e]]#intensity columns
df_ratio = df_iBAQ.copy()
for i in range(df_ratio.shape[1]):
    df_ratio.iloc[:,i] = df_intensity.iloc[:,i] / df_iBAQ.iloc[:,i] #intensity/iBAQ
df_ratioGood = df_ratio.dropna()
df_ratioGood.head()
````

## get  normalized iBAQs for 4 groups
20190209

### notes

```
Segment 3, all samples
1.  All of the samples had a predominant protein band in them (see attached images).  
1a.  This band was used to define the cut borders for "gel segment 3;"  in our lane excisions, "segment 3" was cut just above and just below this band, so that the segment contained only the band.
1b.  This band (now known as segment 3) was so abundant that is was necessary to dilute it 1:20 prior to running it on the LC column.
1c.  However, the other lane fractions (segments #1, #2, #4, #5, #6, #7,and #8) did not require this dilution, and diluting them would have greatly diminished your sensitivity.  
1d.  Thus, they were injected without further dilution.
1e.  Summarizing, segment 3 for each sample was diluted 1:20 prior to injection.
 
2.  For LFQ analysis, this doesn't require any correction;  LFQ strives for relative expression between samples.  

3.  For iBAQ, however, the iBAQ intensities should probably multiplied 10X.  This is an imperfect solution (due to the finer points of electrospray physics) but it's a solution.
Adult samples
1. The adult samples have a similar problem, in that there are 3 other gel regions / bands that were too abundant to load without dilution:  segments #1, #2, #7, and #8.  Thus, they (and segment #3) were diluted 1:20 before injection
2.  These could be handled as proposed above for iBAQ. 

3.  Alternatively, you could do your quantitative thinking based on the band images, that is, the gels clearly show that the adults express proteins that are not very abundant in larvae, and MS was used to ID these proteins.

4.  It's not clear to me how you might best perform LFQ analyses of adult segments.
```
